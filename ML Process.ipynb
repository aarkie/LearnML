Generating random data and Testing the model on a basic ML Process

f(x,z) = 13*xs + 7*zs - 12

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Generating random inputs

```python
observation = 1000
```

```python
xs = np.random.uniform(-10, 10, size=(observation, 1))
zs = np.random.uniform(-10, 10, size = (observation,1))
```

```python
inputs = np.column_stack((xs,zs))
```

```python
print(inputs.shape)
```

```python
noise = np.random.uniform(-1, 1, (observation,1))
```

```python
targets = 13*xs + 7*zs - 12 + noise
```

```python
print(targets)
print(targets.shape)
```

```python
init_range = 0.1
weights = np.random.uniform(low=-init_range, high=init_range, size=(2, 1))
biases = np.random.uniform(low=-init_range, high=init_range, size=1)

print (weights)
print (biases)
```

```python
eta=0.05
```

```python

# optimization algorithm
# gradient descent

for i in range(200):
    
    outputs = np.dot(inputs, weights) + biases
    
    deltas = outputs - targets
    
    loss = (np.sum(targets - outputs)**2)/2 / observation
    delta_scaled = deltas/observation
    weights = weights - eta * np.dot(inputs.T, delta_scaled)
    biases = biases - eta * np.sum(delta_scaled)
    
    print (loss)
    
```
